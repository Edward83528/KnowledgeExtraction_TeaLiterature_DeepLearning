```python
ner_foder="data_ner/"
nerdata=ner_foder+"train.tsv" # 原始檔案
nerdata2=ner_foder+"train2.tsv"
nerdata3=ner_foder+"train3.tsv"

nerdata4_train=ner_foder+"train4.tsv"
nerdata4_dev=ner_foder+"dev4.tsv"
nerdata4_test=ner_foder+"test4.tsv"

nerdata5_train=ner_foder+"train5.tsv"
nerdata5_dev=ner_foder+"dev5.tsv"
nerdata5_test=ner_foder+"test5.tsv"
```

```python
listtemp = [];
end = ['。', 'O']
emptycount=0
import csv

with open(nerdata2, 'w') as f:
    tsv_w = csv.writer(f, delimiter='\t')
    with open(nerdata) as f2:
        tsvreader = csv.reader(f2, delimiter='\t')
        for line in tsvreader:
            if len(line)>0:
              #print(line)
              listtemp.append(line);
              listtemp.append("||||");
              if line[0]=='。':
                tsv_w.writerow(listtemp) # 單行寫入
                listtemp = [];
            else:
              emptycount=emptycount+1
              if emptycount>3:
                tsv_w.writerow(end)
                emptycount=0
```

```python
# 去除空白行
with open(nerdata3, 'w') as f:
    tsv_w = csv.writer(f, delimiter='\t')
    with open(nerdata2) as f2:
        tsvreader = csv.reader(f2, delimiter='\t')
        for line in tsvreader:
          #print(line)
          if line[0]!='。':
            tsv_w.writerow(line) # 單行寫入
```

```python
from sklearn.model_selection import train_test_split
import random

def split(fname, train_ratio, var_ratio):
    lines = fname.readlines()
    n_total = len(lines) # 獲取數據集的總長度

    train_offset = int(n_total * train_ratio)
    val_offset = int(n_total * (train_ratio + var_ratio))
    #print(type(lines))
    random.shuffle(lines) # 按行打亂順序

    train_data = open(nerdata4_train, 'wb')
    val_data = open(nerdata4_dev, 'wb')
    test_data = open(nerdata4_test, 'wb')


    for i, line in enumerate(lines):
        if i < train_offset:
            train_data.write(line)
        elif i < val_offset:
            val_data.write(line)
        else:
            test_data.write(line)

    train_data.close()
    val_data.close()
    test_data.close()
```

```python
fname = open(nerdata3, "rb")
split(fname, train_ratio = 0.8, var_ratio = 0.2)
fname.close()
```
