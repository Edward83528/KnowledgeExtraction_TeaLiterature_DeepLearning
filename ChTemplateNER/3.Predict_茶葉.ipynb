```python
import torch
#cuda是否可用
torch.cuda.is_available()
```

```python
modelDir = 'outputs'
verifyFile = 'data_verify/茶業改良場.json'
entityDir = modelDir+'/best_model_entity_tea_996'
relationDir = modelDir+'/best_model_relation_tea_400'
relationDir50 = modelDir+'/best_model_relation_tea_400_50length'
relationDir100 = modelDir+'/best_model_relation_tea_400_100length'
relationDir200 = modelDir+'/best_model_relation_tea_400_200length'
relationDir300 = modelDir+'/best_model_relation_tea_400_300length'
relationDir400 = modelDir+'/best_model_relation_tea_400_400length'
relationDir500 = modelDir+'/best_model_relation_tea_400基底_加上400length再訓練一次'
tokenizerModel = "fnlp/bart-base-chinese"
```

```python
from transformers import BertTokenizer, BartForConditionalGeneration, Text2TextGenerationPipeline
tokenizer = BertTokenizer.from_pretrained(tokenizerModel)

model_entity = BartForConditionalGeneration.from_pretrained(entityDir)
text2text_generator_entity = Text2TextGenerationPipeline(model_entity, tokenizer)

model_relation = BartForConditionalGeneration.from_pretrained(relationDir)
text2text_generator_relation = Text2TextGenerationPipeline(model_relation, tokenizer)

model_relation50 = BartForConditionalGeneration.from_pretrained(relationDir50)
text2text_generator_relation50 = Text2TextGenerationPipeline(model_relation50, tokenizer)

model_relation100 = BartForConditionalGeneration.from_pretrained(relationDir100)
text2text_generator_relation100 = Text2TextGenerationPipeline(model_relation100, tokenizer)

model_relation200 = BartForConditionalGeneration.from_pretrained(relationDir200)
text2text_generator_relation200 = Text2TextGenerationPipeline(model_relation200, tokenizer)

model_relation300 = BartForConditionalGeneration.from_pretrained(relationDir300)
text2text_generator_relation300 = Text2TextGenerationPipeline(model_relation300, tokenizer)

# 400的現在看起來是最好的
model_relation400 = BartForConditionalGeneration.from_pretrained(relationDir400)
text2text_generator_relation400 = Text2TextGenerationPipeline(model_relation400, tokenizer)

model_relation500 = BartForConditionalGeneration.from_pretrained(relationDir500)
text2text_generator_relation500 = Text2TextGenerationPipeline(model_relation500, tokenizer)
```

```python
detail="為了提高鮮葉產量，延長高產年限，抑制生殖生長，應重施氮肥，配施磷、鉀肥。"
#detail="青心烏龍別名：青心、烏龍、種仔、種茶、軟枝、正叢,外觀：枝葉茂密，茶芽綠黃，葉片為濃綠色，葉形為細長橢圓形，鋸齒細而密，主脈與側脈為小角度銳角"
#detail="台茶十八號，又稱紅玉，屬於大葉種茶樹品種，由台灣野生山茶與緬甸大葉種茶葉雜交育種而成，由台灣農委會茶葉改良場魚池分所培育出來，在1999年發表[1]。主要種植於台灣南投縣魚池鄉，用於製作紅茶，其特色為茶湯深紅，有肉桂香及薄荷香氣。帶有獨特薄荷肉桂香及甜甜的焦糖味。"
#detail="台茶12號母本台農8號+父本硬枝紅心之雜交後裔，試驗代號為2027，茶改場成功育種的排列順序第十二號。於民國70年命名，樹型橫張型，葉橢圓形，葉肉厚，葉色濃綠富光澤，芽綠中帶紫，富茸毛，芽密度高，樹勢強，高產，採摘期長，抗枝枯病。台茶12號的採收量比青心大冇和青心烏龍高20%至50%。金萱品質特徵具一股濃濃的天然「奶香」。"
```
